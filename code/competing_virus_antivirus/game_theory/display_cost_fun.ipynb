# Affichage de la fonction de coût pour le premier jeu pour lequel il n'y pas de coût pour le virus

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('res_simu/big_graph_unicast_virus_parallel3_beta.csv')
df2 = pd.read_csv('res_simu/big_graph_unicast_virus_parallel9_beta.csv')
```

```python
df1 = df1[0:-1:10]
```

```python
df = pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
K = df[['K']]
Y = df[['cost fun']]
EP = df[['epidemic peak']]
```

```python
plt.plot(K,Y, 'blue')
#plt.plot(K,EP, 'red')
plt.legend({"epidemic peak", "cost function"})
plt.xlabel("Cost K (thousands €)")
plt.ylabel("Cost function")
plt.title("Evolution of the cost function")
plt.savefig("foo.png")
plt.show()
```
